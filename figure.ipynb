```python
import matplotlib.pyplot as plt
```

```python
name_list = ["GPT3.5 Naive", "GPT3.5 RAG", "GPT3.5 RAG+Rule", "GPT3.5 RAG+Rule+AutoCoT", 
             "GPT4 Naive", "GPT4 RAG", "GPT4 RAG+Rule", "GPT4 RAG+Rule+AutoCoT",
             "Vicuna7B Naive", "Vicuna7B RAG", "Vicuna7B RAG+Rule", "Vicuna7B RAG+Rule+AutoCoT",
             "Vicuna13B Naive", "Vicuna13B RAG", "Vicuna13B RAG+Rule", "Vicuna13B RAG+Rule+AutoCoT",
             "ChemLLM7B Naive", "ChemDFM13B Naive"
            ]
time_list = [4.89, 4.8, 7.02, 13.97,
             28.87, 29.09, 30.9, 26.7,
             21.39, 28.44, 25.59, 25.79,
             36.74, 62.36, 74.43, 75,
             15.38, 21.45
            ]

acc_list = [0.14, 0.3, 0.36, 0.24, 
            0.24, 0.26, 0.28, 0.26, 
            0.02, 0.06, 0.04, 0.04,
            0.06, 0.06, 0.02, 0,
            0.02, 0.14
           ]


model_category = ["gpt"]*8 + ["vicuna"]*8 + ["chemical"]*2
model_marker_mapping = {
    "gpt": "o",
    "vicuna": "^",
    "chemical": "X"
}

method_category = ["naive", "rag", "rag+rule", "rag+rule+autocot"]*4 + ["naive"]*2
method_color_mapping = {
    "naive": "salmon",
    "rag": "limegreen",
    "rag+rule": "royalblue",
    "rag+rule+autocot": "darkviolet"
    
}
size_category = ["small"]*4 + ["large"]*4 + ["small"]*4 + ["large"]*4 + ["small" , "large"]
size_size_mapping ={"small": 20, "large": 100}
```

```python
fig, axs = plt.subplots(1, 1, figsize=(7,7))
axs.set_xlabel('Avg Running Time', fontsize=14)
axs.set_ylabel('Top10 Accuracy', fontsize=14)
for i, (time, acc) in enumerate(zip(time_list, acc_list)):
    marker = model_marker_mapping[model_category[i]]
    color = method_color_mapping[method_category[i]]
    size = size_size_mapping[size_category[i]]
    axs.scatter(time, acc, marker=marker, c=color, s=size, label=name_list[i])
axs.set_xlim(80, 0)
axs.legend()

fig.savefig('results.png')
```
